# Indexing

```python
import pandas as pd
oo=pd.read_csv('pokemon_data.csv')
oo.head()
```

# Index

```python
#Index(column)- it helps us to access a row or column using a label
type(oo.index)
```

```python
oo.index[100]
```

# set_index()

```python
#It helps us to set our index (Column by switching with the first one)
oo.set_index('HP',inplace=True)
```

```python
#By using inplace, it changes the index in the main dataframe also
oo.head()
```

```python
#second way of using set index
oo.reset_index(inplace=True)
name=oo.set_index('Name')
name.head()
```

# reset_index()

```python
#It is opposite of setting an index by returning the dataframe to default integer based index
oo.reset_index(inplace=True)
oo.head()
```

# sort_index()

```python
#We will be first changing index and then sorting the dataframe alphabetically or nummerically according to the index 
oo.head()
```

```python
oo.set_index('Speed',inplace=True)
oo.head()
```

```python
#sorting according to the index
oo.sort_index(inplace=True)
oo.head()
```

```python
#Sorting in descending order
oo.sort_index(inplace=True,ascending=False)
oo.head()
```

# loc[...]

```python
#It is a label based index and it will give errors when items are not found
#We will have to first set the index for the keyword we choose
#The other method of boolean indexing is much more easier and better
oo.set_index('HP',inplace=True)
```

```python
oo.loc[1]
```

# iloc[...]

```python
#We are doing selection by integer index
#It allows traditional pythonic slicing
oo.iloc[500]
```

```python
oo.iloc[[2,34,66,87,150]]
```

```python
oo.iloc[1:4]
```
