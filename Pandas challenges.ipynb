# Chapter 8 Challenge

```python
d=oo[(oo.NOC=='USA')&(oo.Medal=='Gold')]
d
```

```python
a=d.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0)
a.plot();
```

```python
b=oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
c=b.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head(5).plot(kind='bar');
c
```

# Final Challenge

```python
a=oo[oo.NOC=='USA']
a=oo.Athlete.count()
b=a[['Edition','Athlete','Discipline','Medal']]
b=oo.groupby(['Edition','Athlete','Medal','Discipline']).size().unstack('Medal',fill_value=0)
b['Total']= b['Gold'] + b['Silver'] + b['Bronze']
b.sort_values(['Total','Edition','Athlete','Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
```
