# Basic Analysis

```python
import pandas as pd
oo=pd.read_csv('pokemon_data.csv')
oo.head()
```

# value_counts()

```python
#value_counts()-this is used to count the number of times any item is repeated in a row
oo.Generation.value_counts()
```

```python
#ascending-this will arrange the outputs in ascending order
oo.Legendary.value_counts(ascending=True)
```

```python
#dropna- this will show more clearer data by ignoring the empty cells
oo.Generation.value_counts(ascending=True,dropna=False)
```

# sort_values()

```python
#sorting-it is used to sort the data in ascending or descending order
name=oo.Name.sort_values()
#assigning the sorted data to a variable called name and then callind it
name
```

```python
#In this, there are two sorting, first the data will be sorted if the card is card is legendary or not, the it will be sorted alphabetically with the names
oo.sort_values(by=['Legendary','Name'])
```

# Boolean Indexing

```python
#boolean vectors are of three types- AND-&, OR-| and NOT-~
oo.head()
```

```python
oo.Speed==45
```

```python
#This will give the list of pokemon who have speed 45
oo[oo.Speed==45]
```

```python
oo[(oo.Speed==45)&(oo.HP==45)]
```
