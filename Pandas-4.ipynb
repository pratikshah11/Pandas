# Basic Plotting

```python
import pandas as pd
oo=pd.read_csv('pokemon_data.csv')
o=oo.head(10)
o
```


# Matplotlib

```python
#Matplotlib- it is used in plotting, making graphs etc
#importing matplotlib
import matplotlib.pyplot as plt
#The second line is important is important because it helps to use matplotlib and its also called the magic line
%matplotlib inline
```

# Plot types

```python
#There are 4 types of charts-
#1.Line graph- this like a normal graph comparing two quatities over something
#2.bar chart- this is a bar chart 
#3.barh chart- this is a horizontal bar chart
#4.pie chart- used to compare parts of a whole
```

```python
o.Speed.plot();
#The default plot is line
```

```python
o.Speed.plot(kind='bar');
```

```python
o.Speed.plot(kind='barh');
```

```python
o.Speed.plot(kind='pie');
```

# Plot colors

```python
#You can change the color of the graph also by using this command
o.Speed.plot(kind='bar',color='red');
```


# Figure Size

```python
#we can specify the figure size by setting the height and width 
```

```python
o.Speed.plot(kind='line',figsize=(10,3));
```

# Colormaps

```python
#1.Sequential- The color gets darker from lighter
#2.Diverging- It is used when the information is deviating around a middle value and the colors of other two ends usually being different
#3.Qualitative- The colors are often miscellaneous and there is no relationship 
```

```python
o.Speed.plot(kind='pie',colormap='Pastel1');
```
